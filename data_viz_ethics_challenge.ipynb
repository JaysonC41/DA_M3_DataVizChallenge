```python
plt.figure(figsize=(10,6))

# Aggregate by year
coffee_df["Year"] = coffee_df["Year"].astype(int)

yearly = coffee_df.groupby("Year")[["Sales", "Profit"]].sum().reset_index()


# Sales line
plt.plot(
    yearly["Year"],
    yearly["Sales"],
    marker="o",
    linewidth=2,
    label="Sales",
    color="steelblue"
)

# Profit line
plt.plot(
    yearly["Year"],
    yearly["Profit"],
    marker="o",
    linewidth=2,
    label="Profit",
    color="darkorange"
)

# Titles and labels (neutral + descriptive)
plt.title("Annual Sales and Profit Trends", fontsize=16, fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Total Amount")

# Clean formatting
plt.legend()
plt.grid(alpha=0.3)

# Fix scientific notation if it appears
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(yearly["Year"].astype(int))
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    coffee_df["Sales"],
    coffee_df["Profit"],
    alpha=0.6,
    color="steelblue"
)

plt.title("Relationship Between Sales and Profit", fontsize=16, fontweight="bold")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.grid(alpha=0.3)

plt.show()
```

## About this Graph
This visualization shows how total sales and profit evolve over time. Both metrics appear to move in a similar direction, suggesting that profitability scales with sales volume in this dataset

```python
plt.figure(figsize=(10,6))

avg_sales = coffee_df.groupby("Region")["Sales"].mean().reset_index()

plt.bar(
    avg_sales["Region"],
    avg_sales["Sales"],
    color="steelblue"
)

plt.title("Average Sales by Region", fontsize=16, fontweight="bold")
plt.xlabel("Region")
plt.ylabel("Average Sales")

plt.grid(axis="y", alpha=0.3)

plt.xticks(rotation=45)
plt.tight_layout
plt.show()
```

```python
plt.figure(figsize=(10,6))

import seaborn as sns

sns.boxplot(
    data=coffee_df,
    x="Region",
    y="Sales"
)

plt.title("Sales Distribution by Region", fontsize=16, fontweight="bold")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.grid(axis="y", alpha=0.3)

plt.show()
```

### Breif Explanation ##
To explore the dataset, I used two complementary visual approaches to understand both overall trends and category differences. The line chart shows how total sales and profit change over time, revealing general growth patterns and allowing comparison between revenue and profitability trends. To better understand the relationship between these two variables, the scatter plot highlights how sales and profit are related at the individual data level and helps identify whether higher sales consistently translate into higher profit. For regional analysis, the bar chart compares average sales across regions to identify which areas perform strongest on average. Finally, the box plot adds context by showing the distribution of sales within each region, revealing whether performance is consistent or driven by outliers. Together, these four visualizations provide a balanced and ethical view of the data by showing trends, relationships, comparisons, and variability without distortion.
